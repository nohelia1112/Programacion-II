```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.cvtColor(cv2.imread("/content/COLOMBIA.png"), cv2.COLOR_BGR2RGB)
plt.imshow(imagen)

lugar = ['Los Santos, Santander', 'Mesetas, Meta', 'Los Santos, Santander', 'Zapatoca, Santander', 'Íquira, Huila', 'Colombia-Ecuador, Region Fronteriza', 'Guaca, Santander']
longitud = np.array([73.12, 74.295, 73.169, 73.335, 75.743, 77.931, 72.761])
latitud = np.array([6.792, 3.46, 6.789, 6.767, 2.761, 0.804, 6.978])
profundidad = np.array([134.45, 12.11, 143, 122.1, 24, 3.81, 3])
magnitud = np.array([1.9, 1.2, 1.4, 1.7, 2, 4.3, 1])*20

#print(imagen.shape) #(1045, 765, 3)

#El 66 y el 4 hacen parte de las medidas reales de la longitud y latitud del mapa de Colombia
restaMochitoX = 14-(longitud-66)
X = (restaMochitoX*765)/14
restaMochitoY = 16-(latitud+4)
Y = (restaMochitoY*1045)/16

plt.title("Algunos sismos ocurridos durante los primeros 15 días de abril/2023")
plt.xlabel('Longitud [°]')
plt.ylabel('Latitud [°]')

plt.scatter(X, Y, s=magnitud, c=profundidad, cmap='winter')
plt.colorbar(label='Profundidad')
plt.show()
```
